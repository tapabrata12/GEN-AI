# Agents with multiple tools

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_groq import ChatGroq
from langchain_tavily import TavilySearch
from dotenv import load_dotenv

load_dotenv()
```

```python
web_search_tool = TavilySearch(
    max_results=1,
    topic='general',
    )
```

```python
@tool('calculator',description='Used to perform calculations')

def calculator(expression:str)->str:
    """ Use to perform any mathamatical operation """
    return str(eval(expression))
```

```python
model = ChatGroq(
    model="openai/gpt-oss-120b",
    temperature=2,
    max_tokens=None,
    reasoning_format="parsed",
    timeout=None,
    max_retries=2,
)

agent = create_agent(model=model, tools=[web_search_tool,calculator])
```

```python
user_query = "What is the latest news in AI? and also calculate 2 + 2"

for chunk in agent.stream(
    {"messages": [{"role": "user", "content": user_query}]},
    stream_mode='values',
):
    chunk['messages'][-1].pretty_print()
```
